```python
import os
import torch
import matplotlib.pyplot as plt
```

```python
train_dir = 'train'
```

```python
figure, axes = plt.subplots(5, 5, figsize=(10, 10))

for i, ax in enumerate(axes.flat):
    img_path = os.path.join(train_dir, f'img\circle_{i+1}.pt')
    target_path = os.path.join(train_dir, f'target\circle_{i+1}.pt')
    
    img = torch.load(img_path)
    target = torch.load(target_path)
    
    ax.imshow(img.squeeze().numpy(), cmap='gray')
    ax.axis('off')
    ax.set_title(f"Circles: {target.shape[0]}")

plt.tight_layout()
plt.show()
```
